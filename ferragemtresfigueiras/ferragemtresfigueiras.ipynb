```python
from bs4 import BeautifulSoup
import requests

def get_product_info(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Create a BeautifulSoup object with the response content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the product information
    product_name = ""
    price = ""
    description = ""
    stock = ""

    product_title_element = soup.find('h2', class_='product_title entry-title show-product-nav')
    if product_title_element:
        product_name = product_title_element.text.strip()

    price_element = soup.find('span', class_='woocommerce-Price-amount')
    if price_element:
        price = price_element.text.strip()

    description_element = soup.find('div', class_='description woocommerce-product-details__short-description')
    if description_element:
        description = description_element.text.strip()

    stock_element = soup.find('span', class_='stock')
    if stock_element:
        stock = stock_element.text.strip()

    # Create a dictionary to store the product information
    product_info = {
        'product_name': product_name,
        'price': price,
        'description': description,
        'stock': stock
    }

    return product_info

# URL of the product page
url = "https://ferragemtresfigueiras.com.br/produto/taschibra-kit-directled-3000k-br/"

# Fetch the product information
product_info = get_product_info(url)

# Print the product information
print("Product Name:", product_info['product_name'])
print("Price:", product_info['price'])
print("Description:", product_info['description'])
print("Stock:", product_info['stock'])
```
